```python
import requests
from bs4 import BeautifulSoup

headers = {
    'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/117.0.0.0 Safari/537.36' 
}
url = "https://www.reclameaqui.com.br/empresa/prefeitura-fortaleza/lista-reclamacoes/?pagina=1&status=EVALUATED"
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'lxml')
```

```python
# Encontre todos os links (tags 'a')
evaluateds = soup.find_all('div', attrs={'class': 'sc-1pe7b5t-0'})

#evaluateds = soup.find_all('span', attrs={'class': 'sc-1pe7b5t-4'})

# Itere sobre os links e verifique se eles têm um 'h4' como filho
for one in evaluateds:
    if one.a is not None:
        uri = one.a.attrs['href']

    txt = one.find('span', attrs={'class': 'sc-1pe7b5t-4'})
    if txt is not None:        
        evaluated = txt.text

    print(evaluated, ' -> ', uri)
```
