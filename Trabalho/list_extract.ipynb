```python
import requests

headers = {
    'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/117.0.0.0 Safari/537.36' 
}
url = "https://www.reclameaqui.com.br/empresa/prefeitura-fortaleza/lista-reclamacoes/?pagina=2"
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'lxml')
```

```python
# Encontre todos os links (tags 'a')
links = soup.find_all('a')

# Inicialize uma lista para armazenar os links com 'h4' como filho
links_with_h4 = []

# Itere sobre os links e verifique se eles tÃªm um 'h4' como filho
for link in links:
    h4 = link.find('h4')
    if h4 and 'title' in h4.attrs:
        links_with_h4.append(link)

# Obtenha os atributos 'href' dos links
for link in links_with_h4:
    print("Link:", link['href'])
```
